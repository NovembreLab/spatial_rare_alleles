```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Define the standard deviations
sigma_x = 1
sigma_y = 1

# Define the 2D Gaussian function
Z = np.exp(-(X**2 / (2 * sigma_x**2) + Y**2 / (2 * sigma_y**2)))

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add a color bar
# fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5, label='Gaussian Value')

# Set labels and title
ax.set_title('3D Surface Plot of 2D Gaussian with Different Standard Deviations')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Gaussian Value')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def periodic_gaussian(X, Y, w, L):
    X = np.mod(X + L / 2, L) - L / 2
    Y = np.mod(Y + L / 2, L) - L / 2
    return np.exp(-(X**2 / (2 * w**2) + Y**2 / (2 * w**2)))

L = 50

x = np.linspace(-L/2, L/2, 1000)
y = np.linspace(-L/2, L/2, 1000)
X, Y = np.meshgrid(x, y)

fig = plt.figure(figsize=(19, 6))

widths = [L / 10, L / 2, L, 4 * L]

for i, w in enumerate(widths, start=1):
    Z = periodic_gaussian(X, Y, w, L)
    ax = fig.add_subplot(1, 4, i, projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
    # ax.set_title(r'$w={:.0f}$'.format(w), fontsize=14, loc='center')
    ax.set_zlim(0, 1)
    ax.set_xlim(-peak_range, peak_range)
    ax.set_ylim(-peak_range, peak_range)
    ax.set_xlabel('Spatial Dimension 1', labelpad=-10)
    ax.set_ylabel('Spatial Dimension 2', labelpad=-10)
    ax.set_zlabel('Sampling Probability', labelpad=-10)
    ax.set_zticklabels([])
    ax.set_xticklabels([])
    ax.set_yticklabels([])

    if i==1:
        ax.set_title(r'$w=L/10$',fontsize=14,loc='center')
    if i==2:
        ax.set_title(r'$w=L/2$',fontsize=14,loc='center')
    if i==3:
        ax.set_title(r'$w=L$',fontsize=14,loc='center')
    if i==4:
        ax.set_title(r'$w=4L$',fontsize=14,loc='center')

plt.tight_layout()
plt.savefig('sampling_ex_periodic.pdf', bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1, 4, figsize=(19, 6))

x = np.linspace(-50, 50, 1000)
y = np.linspace(-50, 50, 1000)
X, Y = np.meshgrid(x, y)

L = 50

def gaussian_norm_factor(w):
    return 1 / (2 * np.pi * w**2)

# Plot A (L/10)
w = L/10
Z = gaussian_norm_factor(w) * np.exp(-(X**2 / (2 * w**2) + Y**2 / (2 * w**2)))
im1 = axs[0].imshow(Z, extent=(-50, 50, -50, 50), origin='lower', cmap='viridis')
axs[0].set_title(r'$w$=L/10')
axs[0].set_xlabel('Spatial Dimension 1')
axs[0].set_ylabel('Spatial Dimension 2')
cbar1 = plt.colorbar(im1, ax=axs[0], fraction=0.046, pad=0.04)
cbar1.set_label('Sampling Probability')

# Plot B (L/2)
w = L/2
Z = gaussian_norm_factor(w) * np.exp(-(X**2 / (2 * w**2) + Y**2 / (2 * w**2)))
im2 = axs[1].imshow(Z, extent=(-50, 50, -50, 50), origin='lower', cmap='viridis')
axs[1].set_title(r'$w$=L/2')
axs[1].set_xlabel('Spatial Dimension 1')
axs[1].set_ylabel('Spatial Dimension 2')
cbar2 = plt.colorbar(im2, ax=axs[1], fraction=0.046, pad=0.04)
cbar2.set_label('Sampling Probability')

# Plot C (L)
w = L
Z = gaussian_norm_factor(w) * np.exp(-(X**2 / (2 * w**2) + Y**2 / (2 * w**2)))
im3 = axs[2].imshow(Z, extent=(-50, 50, -50, 50), origin='lower', cmap='viridis')
axs[2].set_title(r'$w$=L')
axs[2].set_xlabel('Spatial Dimension 1')
axs[2].set_ylabel('Spatial Dimension 2')
cbar3 = plt.colorbar(im3, ax=axs[2], fraction=0.046, pad=0.04)
cbar3.set_label('Sampling Probability')

# Plot D (4L)
w = 10*L
Z = gaussian_norm_factor(w) * np.exp(-(X**2 / (2 * w**2) + Y**2 / (2 * w**2)))
im4 = axs[3].imshow(Z, extent=(-50, 50, -50, 50), origin='lower', cmap='viridis')
axs[3].set_title(r'$w$=4L')
axs[3].set_xlabel('Spatial Dimension 1')
axs[3].set_ylabel('Spatial Dimension 2')
cbar4 = plt.colorbar(im4, ax=axs[3], fraction=0.046, pad=0.04)
cbar4.set_label('Sampling Probability')

# Adding ABCD labels
axs[0].text(-0.1, 1.1, 'A', transform=axs[0].transAxes,
            fontsize=22, fontweight='bold', va='top', ha='right')
axs[1].text(-0.1, 1.1, 'B', transform=axs[1].transAxes,
            fontsize=22, fontweight='bold', va='top', ha='right')
axs[2].text(-0.1, 1.1, 'C', transform=axs[2].transAxes,
            fontsize=22, fontweight='bold', va='top', ha='right')
axs[3].text(-0.1, 1.1, 'D', transform=axs[3].transAxes,
            fontsize=22, fontweight='bold', va='top', ha='right')

plt.tight_layout()
# plt.savefig('sampling_ex_2D_no_normalization.pdf', bbox_inches='tight')
plt.savefig('sampling_ex_2D.pdf')
```

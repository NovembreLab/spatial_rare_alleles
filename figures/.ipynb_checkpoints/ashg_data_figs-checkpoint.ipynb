```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.colors import ListedColormap
```

```python
# function to map to bin
def map_to_bin(x,y,xedges,yedges):
    return [np.digitize(x,xedges)-1,np.digitize(y,yedges)-1]

# function to get bin center
def get_bin_center(bin_idx,xedges,yedges):
    x_idx, y_idx = bin_idx
    center_x = (xedges[x_idx] + xedges[x_idx + 1]) / 2
    center_y = (yedges[y_idx] + yedges[y_idx + 1]) / 2
    return [center_x, center_y]
```

```python
df = pd.read_csv('../metadata/metadata_cleaned_wes.csv')
df_geo = df[(df['birth_east_coord'].notna()) & (df['birth_UKelsewhere']!='Elsewhere') & (df['used_in_pca']==1) & (df['within_1epsilon_pca']==True)]
df_pca = df[(df['used_in_pca']==1)]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
center = 'centerE16N4'
w_list = ['50000', '100000', '150000']
fig, ax = plt.subplots(1, 2,figsize=(12,6))
colors=['#5E2B81', '#3E7B98', '#A0C700']
vartypes=['lof','nolof']
for i,vt in enumerate(vartypes):
    # Load uniform data
    sfs_unif = pd.read_csv(f'../results/sfs/chr1_{vt}_uniformgeo_nSIR10000_nSIRreps10.SIRsfs', sep=' ')
    sfs_unif_grouped = sfs_unif.groupby('OBS_N')['COUNT_SITES'].agg(['mean', 'std'])
    sfs_unif_avg = sfs_unif_grouped['mean']
    sfs_unif_std = sfs_unif_grouped['std']
    
    # Create log bins starting from OBS_N=10 for consistent binning
    log_bins = np.logspace(np.log10(10), np.log10(sfs_unif_avg.index.max()), num=50)
    log_bins = np.unique(np.round(log_bins))  # Ensure bins are unique and rounded
    
    # Plot uniform data
    # Raw data for OBS_N < 10
    obs_n_uniform = sfs_unif_avg.index[sfs_unif_avg.index < 10]
    count_sites_uniform = sfs_unif_avg.values[sfs_unif_avg.index < 10]
    std_uniform = sfs_unif_std.values[sfs_unif_avg.index < 10]
    
    # Binning for uniform data for values >= 10
    sfs_unif_binned, _ = np.histogram(sfs_unif_avg.index[sfs_unif_avg.index >= 10], 
                                       bins=log_bins, 
                                       weights=sfs_unif_avg.values[sfs_unif_avg.index >= 10])
    std_unif_binned, _ = np.histogram(sfs_unif_avg.index[sfs_unif_avg.index >= 10], 
                                       bins=log_bins, 
                                       weights=sfs_unif_std.values[sfs_unif_avg.index >= 10])
    
    # Combine raw and binned data for Uniform
    obs_n_uniform_combined = np.concatenate([obs_n_uniform, log_bins[:-1]])
    count_sites_uniform_combined = np.concatenate([count_sites_uniform, sfs_unif_binned])
    std_uniform_combined = np.concatenate([std_uniform, std_unif_binned])

    # Plot combined data for Uniform with error bars
    ax[i].errorbar(obs_n_uniform_combined, count_sites_uniform_combined, 
                yerr=std_uniform_combined, label='Uniform', linestyle='-', 
                marker='o', capsize=5, elinewidth=1, fmt='o',color='darkgray')
    
    # Plot Gaussian data
    for j,w in enumerate(w_list):
        sfs = pd.read_csv(f'../results/sfs/chr1_{vt}_{center}geo{w}_nSIR10000_nSIRreps10.SIRsfs', delimiter=' ')
        sfs_avg_grouped = sfs.groupby('OBS_N')['COUNT_SITES'].agg(['mean', 'std'])
        sfs_avg = sfs_avg_grouped['mean']
        sfs_std = sfs_avg_grouped['std']
    
        # Raw data for Gaussian for values < 10
        obs_n_gaussian = sfs_avg.index[sfs_avg.index < 10]
        count_sites_gaussian = sfs_avg.values[sfs_avg.index < 10]
        std_gaussian = sfs_std.values[sfs_avg.index < 10]
    
        # Binning for Gaussian data for values >= 10
        sfs_avg_binned, _ = np.histogram(sfs_avg.index[sfs_avg.index >= 10], 
                                          bins=log_bins, 
                                          weights=sfs_avg.values[sfs_avg.index >= 10])
        std_gaussian_binned, _ = np.histogram(sfs_avg.index[sfs_avg.index >= 10], 
                                               bins=log_bins, 
                                               weights=sfs_std.values[sfs_avg.index >= 10])
    
        # Combine raw and binned data for Gaussian
        obs_n_gaussian_combined = np.concatenate([obs_n_gaussian, log_bins[:-1]])
        count_sites_gaussian_combined = np.concatenate([count_sites_gaussian, sfs_avg_binned])
        std_gaussian_combined = np.concatenate([std_gaussian, std_gaussian_binned])
    
        # Plot combined data for Gaussian with error bars
        ax[i].errorbar(obs_n_gaussian_combined, count_sites_gaussian_combined, 
                    yerr=std_gaussian_combined, 
                    label=f'Gaussian (width: {int(int(w)/1000)} km)', 
                    linestyle='-', marker='o', capsize=5, elinewidth=1, fmt='o',color=colors[j])
    
    # Add legend and labels
    ax[i].set_yscale('log')
    ax[i].set_xscale('log')
    ax[i].set_ylim(5e-1, 1e4)
    if i==1:
        ax[i].legend(frameon=False,fontsize=12)
    ax[i].set_xlabel('Allele count',fontsize=14)
    ax[i].set_ylabel('Number of variants',fontsize=14)
    ax[i].set_xlim(1e0, 1e2)
    if vt=='lof':
        ax[i].set_title('LoF variants, chromosome 1',loc='left',fontstyle='italic',fontsize=16)
    else:
        ax[i].set_title('Non-LoF variants (subset), chromosome 1',loc='left',fontstyle='italic',fontsize=16)
plt.tight_layout()
# plt.show()
plt.savefig('geo_sfs.png',dpi=600)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
reps = np.arange(10) + 1
colors = ['steelblue', 'darkorange']  # Colors for LOF and non-LOF

# Use only the E16N4 center
center = 'centerE16N4'
w_list = ['50000','100000','150000']
# Create a figure with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows, 2 columns without shared y-axis
row_titles = [
    'Sum of segregating sites',
    'Number of singletons',
    'Average frequency of segregating alleles',
    'Average frequency of all alleles'
]

# Flatten the axes array for easier indexing
axs = axs.flatten()
vartypes=['nolof','lof']
# Loop through each row (statistic)
n = 10000
for row in range(4):
    ax = axs[row]  # Select the current subplot
    ax.set_title(row_titles[row],loc='left',fontstyle='italic',fontsize=16)  # Set title for the statistic
    ax.set_ylabel(row_titles[row])  # Set title for the statistic
    ax.set_xlabel('Width of sampling kernel')

    labels = [f'50 km', f'100 km', f'150 km', 'Uniform']
    
    for vt_idx, vt in enumerate(vartypes):  # Loop through vartypes (LOF and non-LOF)
        means = []
        errors = []

        for idx, w in enumerate(w_list + ['uniform']):
            sfs_file = f'../results/sfs/chr1_{vt}_{center}geo{w}_nSIR10000_nSIRreps10.SIRsfs' if w != 'uniform' else f'../results/sfs/chr1_{vt}_uniformgeo_nSIR10000_nSIRreps10.SIRsfs'
            sfs = pd.read_csv(sfs_file, delimiter=' ')

            if row == 0:
                # Row 1: Sum of Segregating Sites (excluding OBS_N=0)
                segregating_sites = [
                    sfs[(sfs['SIR_REP'] == rep) & (sfs['OBS_N'] > 0)]['COUNT_SITES'].sum()
                    for rep in reps
                ]
                current_mean = np.mean(segregating_sites)
                means.append(current_mean)
                errors.append(np.std(segregating_sites))

            elif row == 1:
                # Row 2: Number of Singletons (OBS_N=1)
                singletons = [
                    sfs[(sfs['SIR_REP'] == rep) & (sfs['OBS_N'] == 1)]['COUNT_SITES'].sum()
                    for rep in reps
                ]
                current_mean = np.mean(singletons)
                means.append(current_mean)
                errors.append(np.std(singletons))

            elif row == 2:
                # Row 3: Average Counts of Segregating Alleles (OBS_N > 0)
                total_count_contributions = []
                for rep in reps:
                    contributions = sfs[(sfs['SIR_REP'] == rep) & (sfs['OBS_N'] > 0)]
                    total_contribution = (contributions['OBS_N'] * contributions['COUNT_SITES']).sum() / contributions['COUNT_SITES'].sum()
                    total_count_contributions.append(total_contribution/n)

                current_mean = np.mean(total_count_contributions)
                means.append(current_mean)
                errors.append(np.std(total_count_contributions))

            elif row == 3:
                # Row 4: Average Counts of All Alleles (OBS_N >= 0)
                total_count_contributions = []
                for rep in reps:
                    contributions = sfs[(sfs['SIR_REP'] == rep)]
                    total_contribution = (contributions['OBS_N'] * contributions['COUNT_SITES']).sum() / contributions['COUNT_SITES'].sum()
                    total_count_contributions.append(total_contribution/n)

                current_mean = np.mean(total_count_contributions)
                means.append(current_mean)
                errors.append(np.std(total_count_contributions))
            if w is 'uniform':
                leftmost = means[0]
                rightmost = means[-1]
                ratio = rightmost / leftmost
                
            
                # if vt=='lof':
                #     ratio_text = f'Uniform/Narrow (LoF): {ratio:.2f}'
                #     ax.text(0.05, 0.1, ratio_text, ha='left', va='center', transform=ax.transAxes, fontsize=12, color='black')
                # else: 
                #     ratio_text = f'Uniform/Narrow (non-LoF): {ratio:.2f}'
                #     ax.text(0.05, 0.2, ratio_text, ha='left', va='center', transform=ax.transAxes, fontsize=12, color='black')
                if row in [1,3]:
                    if vt=='lof':
                        ratio_text = f'Uniform/Narrow (LoF): {ratio:.2f}'
                        ax.text(0.05, 0.9, ratio_text, ha='left', va='center', transform=ax.transAxes, fontsize=12, color='black')
                    else: 
                        ratio_text = f'Uniform/Narrow (non-LoF): {ratio:.2f}'
                        ax.text(0.05, 0.8, ratio_text, ha='left', va='center', transform=ax.transAxes, fontsize=12, color='black')
                else:
                    if vt=='lof':
                        ratio_text = f'Uniform/Narrow (LoF): {ratio:.2f}'
                        ax.text(0.05, 0.2, ratio_text, ha='left', va='center', transform=ax.transAxes, fontsize=12, color='black')
                    else: 
                        ratio_text = f'Uniform/Narrow (non-LoF): {ratio:.2f}'
                        ax.text(0.05, 0.1, ratio_text, ha='left', va='center', transform=ax.transAxes, fontsize=12, color='black')


        # Ensure that the lengths of labels, means, and errors match
        if len(labels) == len(means) == len(errors):
            # Plot bars for the current statistic with error bars
            if vt=='lof':
                vt_lab='LoF'
                vt_col='darkorange'
            else:
                vt_lab='non-LoF'
                vt_col='steelblue'
            ax.errorbar(labels, means, yerr=errors, capsize=10, markersize=10, fmt='o', label=vt_lab,color=vt_col,linestyle='--')
        else:
            print(f"Mismatch in lengths: labels({len(labels)}), means({len(means)}), errors({len(errors)})")

    if row == 0:
        ax.legend(title='Variant Type',loc='lower right',frameon=False,fontsize=14,title_fontsize=14)
    
# Set the common y-label for the first subplot
# axs[0].set_ylabel('Counts')
axs[0].set_yscale('log')
axs[1].set_yscale('log')
axs[2].set_yscale('log')
axs[3].set_yscale('log')
axs[0].set_ylim(1e3,1e4)
axs[1].set_ylim(1e3,1e4)
axs[2].set_ylim(1e-5,1e-1)
axs[3].set_ylim(1e-5,1e-1)
# axs[3].set_ylim(0.9e0,2.2e2)


plt.tight_layout()
plt.savefig('sumstats_geo_poster.png',dpi=600)
```

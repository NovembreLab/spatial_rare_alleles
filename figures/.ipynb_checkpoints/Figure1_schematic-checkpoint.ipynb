```python
def branching_diffusion_with_death(num_particles, steps, branch_prob, death_prob, diff_std, max_particles=10,
                                  smin=-10, smax=-5, new_particle_prob=0.1):
    positions = np.zeros((steps, max_particles))
    alive = np.zeros((steps, max_particles), dtype=bool)
    colors = [
        'red', 'green', 'teal', 'skyblue', 'darkorange', 'darkviolet', 'brown', 'purple', 
        'magenta', 'gold', 'gold', 'lime', 'cyan', 'pink', 'orange', 'brown', 'salmon', 
        'blue', 'red', 'coral', 'navy', 'orchid', 'chartreuse', 'khaki', 'orchid', 'crimson', 
        'yellow', 'indigo', 'turquoise', 'salmon'
    ]

    alive[0, 0] = True

    live_counts = np.zeros(steps)  
    live_counts[0] = 1
    
    live_counts_samp = np.zeros(steps)
    if smin <= 0 <= smax:
        live_counts_samp[0] = 1

    births = []  
    deaths = [] 
    newparticles = []
    
    for step in range(1, steps):
        for i in range(max_particles):
            if np.random.rand() < new_particle_prob and num_particles < max_particles:
                num_particles += 1
                positions[step, num_particles-1] = np.random.uniform(smin, smax)
                alive[step, num_particles-1] = True
                # colors[num_particles-1] = f"C{i}"  # Color based on new particle index
                newparticles.append((step, positions[step, num_particles-1], colors[num_particles-1]))  # Record birth event
        for i in range(num_particles):
            if not alive[step-1, i]:
                continue
            # Diffusion
            move = np.random.normal(0, diff_std)
            positions[step, i] = positions[step-1, i] + move
            # Death 
            if np.random.rand() < death_prob:
                deaths.append((step-1, positions[step-1, i], colors[i]))  # Record death event with the current position
                continue
            else:
                alive[step, i] = True  # If not dead, mark the particle as alive
            # Branching
            if np.random.rand() < branch_prob and num_particles < max_particles:
                num_particles += 1
                positions[step, num_particles-1] = positions[step, i]
                alive[step, num_particles-1] = True
                colors[num_particles-1] = colors[i]  # Inherit parent particle's color
                births.append((step, positions[step, i], colors[num_particles-1]))  # Record birth event
        live_counts[step] = np.sum(alive[step, :])
        
        for i in range(num_particles):
            if alive[step, i] and smin <= positions[step, i] <= smax:
                live_counts_samp[step] += 1
        
    return positions, alive, colors, births, deaths, live_counts, live_counts_samp, newparticles
```

```python
fig = plt.figure(figsize=(12, 9)) 
gs = GridSpec(2, 2, width_ratios=[2, 1])

ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[1, 0])
ax3 = plt.subplot(gs[0, 1])
ax4 = plt.subplot(gs[1, 1])

time_slice = 38
lim=85
sample_lim = 35
num_particles = 1
steps = 50
branch_prob = 0.2
death_prob = 0.01
diff_std = 4.0
seed = 356#np.random.randint(0,1000,1)
smin = -50
smax = 50
new_particle_prob = 0.02
np.random.seed(seed)
print(seed)

positions, alive, colors, births, deaths, live_counts, live_counts_sample, newparticles = branching_diffusion_with_death(num_particles, 
                                                                          steps, branch_prob, 
                                                                          death_prob, diff_std,
                                                                          max_particles=25, smin=smin,
                                                                          smax=smax, new_particle_prob=new_particle_prob)

for i in range(positions.shape[1]):
    alive_indices = np.where(alive[:, i])[0]
    if alive_indices.any():
        for j in range(len(alive_indices)-1):
            if alive_indices[j+1] - alive_indices[j] == 1:
                ax1.plot(range(alive_indices[j], alive_indices[j+1]+1), 
                         positions[alive_indices[j]:alive_indices[j+1]+1, i], 
                         color=colors[i], linewidth=2)

for death in deaths:
    ax1.scatter(death[0], death[1], c=death[2], marker='X', s=60, zorder=3, edgecolors='black')

ax1.scatter(0, 0, marker='D', color=colors[0], s=60, zorder=3, edgecolors='black')  
for part in newparticles:
    ax1.scatter(part[0], part[1], marker='D', color=part[2], s=60, zorder=3, edgecolors='black')    

ax1.set_xlabel("Time")
ax1.set_ylabel("Spatial Dimension 2") 

ax1.set_yticks([0])
ax1.axvline(x=time_slice,linestyle='dashed',linewidth=2,color='black')
ax1.set_ylim(-lim,lim)
ax1.axhspan(-sample_lim, sample_lim, facecolor='gray', alpha=0.2, edgecolor='none')
alive_indices = np.where(alive[time_slice])[0]
y_positions = positions[time_slice, alive_indices]
x_positions = np.random.uniform(-65, 65, size=len(alive_indices))  
ax3.scatter(x_positions, y_positions, color=[colors[i] for i in alive_indices], marker='o', s=100, edgecolors='black')
ax3.set_xlim(-lim,lim)
ax3.set_ylim(-lim,lim)

box = patches.Rectangle((-sample_lim, -sample_lim), 2*sample_lim, 2*sample_lim, linewidth=1, edgecolor='none', facecolor='gray',alpha=0.2)
ax3.add_patch(box)
ax3.set_xlabel("Spatial Dimension 1")
ax3.set_ylabel("Spatial Dimension 2")

num_particles = 1
steps = 50
branch_prob = 0.02
death_prob = 0.04
diff_std = 1
seed = 591#np.random.randint(0,1000,1)
smin = -60
smax = 60
new_particle_prob = 0.02
np.random.seed(seed)
print(seed)

positions, alive, colors, births, deaths, live_counts, live_counts_sample, newparticles = branching_diffusion_with_death(num_particles, 
                                                                          steps, branch_prob, 
                                                                          death_prob, diff_std,
                                                                          max_particles=12, smin=smin,
                                                                          smax=smax, new_particle_prob=new_particle_prob)

for i in range(positions.shape[1]):
    alive_indices = np.where(alive[:, i])[0]
    if alive_indices.any():
        for j in range(len(alive_indices)-1):
            if alive_indices[j+1] - alive_indices[j] == 1:
                ax2.plot(range(alive_indices[j], alive_indices[j+1]+1), 
                         positions[alive_indices[j]:alive_indices[j+1]+1, i], 
                         color=colors[i], linewidth=2)

for death in deaths:
    ax2.scatter(death[0], death[1], c=death[2], marker='X', s=60, zorder=3, edgecolors='black')

ax2.scatter(0, 0, marker='D', color=colors[0], s=60, zorder=3, edgecolors='black')  
for part in newparticles:
    ax2.scatter(part[0], part[1], marker='D', color=part[2], s=60, zorder=3, edgecolors='black')  

ax2.set_xlabel("Time")
ax2.set_ylabel("Spatial Dimension 2")
ax2.set_yticks([0])
ax2.axvline(x=time_slice,linestyle='dashed',linewidth=2,color='black')
ax2.axhspan(-sample_lim, sample_lim, facecolor='gray', alpha=0.2, edgecolor='none')
ax2.set_ylim(-lim,lim)

alive_indices = np.where(alive[time_slice])[0]
y_positions = positions[time_slice, alive_indices]
x_positions = np.random.uniform(-20, 20, size=len(alive_indices))  
ax4.scatter(x_positions, y_positions, color=[colors[i] for i in alive_indices], marker='o', s=100, edgecolors='black')
ax4.set_xlim(-lim,lim)
ax4.set_ylim(-lim,lim)
box = patches.Rectangle((-sample_lim, -sample_lim), 2*sample_lim, 2*sample_lim, linewidth=1, edgecolor='none', facecolor='gray',alpha=0.2)
ax4.add_patch(box)

ax4.set_xlabel("Spatial Dimension 1")
ax4.set_ylabel("Spatial Dimension 2")
for ax in [ax1, ax2, ax3, ax4]:
    ax.set_xticks([])
    ax.set_yticks([])

matplotlib.rcParams['axes.labelsize'] = 18

ax1.set_title('Weak selection', loc='left', fontstyle='italic',fontsize=22)
ax2.set_title('Strong selection', loc='left', fontstyle='italic',fontsize=22)
ax1.text(-0.1, 1.1, 'A', transform=ax1.transAxes, fontsize=24, fontweight='bold', va='top', ha='right')
ax2.text(-0.1, 1.1, 'B', transform=ax2.transAxes, fontsize=24, fontweight='bold', va='top', ha='right')
ax3.legend([box], ['Sampled Region'], loc='lower right', fontsize=12,frameon=True)
ax4.legend([box], ['Sampled Region'], loc='lower right', fontsize=12,frameon=True)
plt.tight_layout(pad=2.0)
plt.savefig("schematic.pdf")
```

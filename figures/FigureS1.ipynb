```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def periodic_gaussian(X, Y, w, L):
    X = np.mod(X + L / 2, L) - L / 2
    Y = np.mod(Y + L / 2, L) - L / 2
    return np.exp(-(X**2 / (2 * w**2) + Y**2 / (2 * w**2)))

L = 50

x = np.linspace(-L/2, L/2, 1000)
y = np.linspace(-L/2, L/2, 1000)
X, Y = np.meshgrid(x, y)

fig = plt.figure(figsize=(19, 6))

widths = [L / 10, L / 2, L, 4 * L]
peak_range=L/2
for i, w in enumerate(widths, start=1):
    Z = periodic_gaussian(X, Y, w, L)
    ax = fig.add_subplot(1, 4, i, projection='3d')
    surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
    # ax.set_title(r'$w={:.0f}$'.format(w), fontsize=14, loc='center')
    ax.set_zlim(0, 1)
    ax.set_xlim(-peak_range, peak_range)
    ax.set_ylim(-peak_range, peak_range)
    ax.set_xlabel('Spatial Dimension 1', labelpad=-10)
    ax.set_ylabel('Spatial Dimension 2', labelpad=-10)
    ax.set_zlabel('Sampling Probability', labelpad=-10)
    ax.set_zticklabels([])
    ax.set_xticklabels([])
    ax.set_yticklabels([])

    if i==1:
        ax.set_title(r'$w=L/10$',fontsize=14,loc='center')
    if i==2:
        ax.set_title(r'$w=L/2$',fontsize=14,loc='center')
    if i==3:
        ax.set_title(r'$w=L$',fontsize=14,loc='center')
    if i==4:
        ax.set_title(r'$w=4L$',fontsize=14,loc='center')

plt.tight_layout()
plt.savefig('sampling_ex_periodic.pdf', bbox_inches='tight')
```

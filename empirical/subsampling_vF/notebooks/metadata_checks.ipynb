```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../metadata/metadata_cleaned_wes.csv')
```

```python
df['closest_10k_pca'].sum()
```

```python
df['closest_10k_geo'].sum()
```

```python
colnames = ['closest_10k_pca','within_1epsilon_pca','within_5epsilon_pca','within_10epsilon_pca','within_50epsilon_pca','within_100epsilon_pca']
for colname in colnames:
    fig,ax = plt.subplots(figsize=(8,8))
    ax.scatter(df['PC1'],df['PC2'],s=0.01,alpha=1,c=df[colname].map({True:'blue',False:'lightgray'}))
    ax.set_title(colname)
    plt.show()
```

```python
colnames = ['closest_10k_geo','within_5epsilon_geo','within_10epsilon_geo','within_50epsilon_geo']
for colname in colnames:
    fig,ax = plt.subplots(figsize=(5,8))
    ax.scatter(df['birth_east_coord'],df['birth_north_coord'],s=0.01,alpha=1,c=df[colname].map({True:'blue',False:'lightgray'}))
    ax.set_title(colname)
    plt.show()
```

```python
testids = pd.read_csv('../data/id_lists/50epsilongeo_seed1_n10000.tsv',delimiter='\t')
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
### load data

# pca
evals = pd.read_csv('../metadata/ukb_pca_plink2.eigenval',delim_whitespace=True,header=None)
evecs = pd.read_csv('../metadata/ukb_pca_plink2.eigenvec',delim_whitespace=True)

# meta
meta = pd.read_csv('../metadata/participant_metadata_all.csv')

# wes ids
wes = pd.read_csv('../metadata/wes_ids_all.txt',header=None)
```

```python
meta
```

```python
# format metadata
meta = meta.rename(columns={'participant.eid':'id'})

# undo codings
meta['participant.p54_i0'] = meta['participant.p54_i0'].astype(str)
center_coding = pd.read_csv("../metadata/codings/coding10.tsv",delimiter="\t",names=["participant.p54_i0","assessment_center"],header=0)
center_dict = dict(zip(center_coding['participant.p54_i0'].astype(str),center_coding['assessment_center']))
meta['assessment_center'] = meta['participant.p54_i0'].map(center_dict)
meta = meta.drop('participant.p54_i0',axis=1)

meta['participant.p1647_i0'] = meta['participant.p1647_i0'].astype(str)
meta['participant.p1647_i0'] = meta['participant.p1647_i0'].str.replace(r'\.0$', '',regex=True)
birth_UKelsewhere_coding = pd.read_csv("../metadata/codings/coding100420.tsv",delimiter="\t",usecols=[0,1],names=["participant.p1647_i0","birth_UKelsewhere"],header=0)
birth_UKelsewhere_dict = dict(zip(birth_UKelsewhere_coding['participant.p1647_i0'].astype(str),birth_UKelsewhere_coding['birth_UKelsewhere']))
meta['birth_UKelsewhere'] = meta['participant.p1647_i0'].map(birth_UKelsewhere_dict)
meta = meta.drop('participant.p1647_i0',axis=1)

# in coordinates, replace -1 with NA
meta = meta.rename(columns={'participant.p130_i0':'birth_east_coord','participant.p129_i0':'birth_north_coord','participant.p22006':'whitebritish','participant.p22020':'used_in_pca'})
meta.loc[meta['birth_east_coord'] == -1, 'birth_east_coord'] = pd.NA
meta.loc[meta['birth_north_coord'] == -1, 'birth_north_coord'] = pd.NA

meta['whitebritish'] = meta['whitebritish'].fillna(0)
meta['used_in_pca'] = meta['used_in_pca'].fillna(0)
```

```python
meta
```

```python
# filter to ids in wes data
wes = wes.rename(columns={wes.columns[0]:'id'})
filtered_meta = pd.merge(meta,wes,on='id')
filtered_meta
```

```python
evecs = evecs.rename(columns={'IID':'id'})
meta_pcs = pd.merge(filtered_meta,evecs.iloc[:,1:],on='id',how='left')
```

```python
#meta_pcs.to_csv('meta_wes_pcs.csv',index=False)
```

```python
meta_pcs
```

```python
plt.scatter(meta_pcs['PC1'],meta_pcs['PC2'],s=0.1)
```

```python
# centroid in 2 dim
centroid = meta_pcs.iloc[:,7:9].median()
centroid
```

```python
distances = np.sqrt(((meta_pcs.iloc[:,7:9] - centroid) ** 2).sum(axis=1))
meta_pcs['distance_to_centroid'] = distances
```

```python
mean_distances = meta_pcs.groupby('whitebritish')['distance_to_centroid'].mean()
mean_distances
```

```python
plt.figure(figsize=(10, 6))
plt.hist(meta_pcs['distance_to_centroid'], bins=40, edgecolor='black')
plt.title('Histogram of Distance to Centroid')
plt.xlabel('Distance to Centroid')
plt.ylabel('Frequency')
plt.show()
```

```python
all_distances = meta_pcs['distance_to_centroid']
bins = np.histogram(all_distances, bins=80)[1]

plt.figure(figsize=(6, 4))
sns.histplot(data=meta_pcs, x='distance_to_centroid', hue='whitebritish', bins=bins, kde=False, palette='Set1')
plt.title('Histogram of Distance to Centroid Stratified by whitebritish')
plt.xlabel('Distance to Centroid')
plt.ylabel('Frequency')

plt.show()
```

```python
epsilon = 0.0001
meta_pcs_samp['within_1epsilon'] = meta_pcs_samp['distance_to_centroid'] < 1 * epsilon
meta_pcs_samp['within_3epsilon'] = meta_pcs_samp['distance_to_centroid'] < 3 * epsilon
meta_pcs_samp['within_5epsilon'] = meta_pcs_samp['distance_to_centroid'] < 5 * epsilon
meta_pcs_samp['within_10epsilon'] = meta_pcs_samp['distance_to_centroid'] < 10 * epsilon
meta_pcs_samp['within_50epsilon'] = meta_pcs_samp['distance_to_centroid'] < 50 * epsilon
meta_pcs_samp['within_100epsilon'] = meta_pcs_samp['distance_to_centroid'] < 100 * epsilon
```

```python
meta_pcs_samp[meta_pcs_samp['within_1epsilon']==True]['birth_UKelsewhere'].value_counts()
```

```python
meta_pcs_samp_noNA = meta_pcs_samp.dropna(subset=['PC1'])
closest_10k = meta_pcs_samp_noNA.nsmallest(10000,'distance_to_centroid')
closest_10k['birth_UKelsewhere'].value_counts()
meta_pcs_samp['closest_10k'] = meta_pcs_samp.index.isin(closest_10k.index)
```

```python
meta_pcs_samp[meta_pcs_samp['closest_10k']==True]['used_in_pca'].sum()
```

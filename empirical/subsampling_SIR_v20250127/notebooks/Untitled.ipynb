```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import glob
import re
```

```python
files = glob.glob('../results/daf/chr1_mis*')
dfs = {file.split('/')[-1]: pd.read_csv(file, delimiter='\t') for file in files}

# Specifically filter for SNPs that are fixed in all iterations in the 'uniformpca' file
uniformpca_file = [file for file in files if 'uniformpca' in file][0]
uniformpca_df = pd.read_csv(uniformpca_file, delimiter='\t')

# Find SNPs that are fixed (Derived_Frequency == 1) in all iterations for uniformpca
fixed_in_uniformpca = uniformpca_df.groupby('POS').filter(lambda x: (x['Derived_Frequency'] == 1).all())['POS']

# Now we will remove those fixed SNPs from all datasets
mean_freqs = {}
for file, df in dfs.items():
    scenario = file.split('_')[2]  # Extract 'geo' or 'pca' from the filename
    print(scenario)
    
    # Exclude SNPs that are fixed in all iterations of 'uniformpca'
    df_filtered = df[~df['POS'].isin(fixed_in_uniformpca)]  # Remove fixed SNPs from uniformpca
    mean_freq = df_filtered.groupby('ITER')['Derived_Frequency'].mean()  # Mean frequency per iteration
    mean_freqs[scenario] = mean_freq

mean_freqs_df = pd.DataFrame(mean_freqs)

means = mean_freqs_df.mean()
std_devs = mean_freqs_df.std()

# Define custom order for plotting
custom_order = [
    'centerE16N4geo50000', 'centerE16N4geo100000', 'centerE16N4geo150000',
    'centerE9N9geo50000', 'centerE9N9geo100000', 'centerE9N9geo150000',
    'centerE6N4geo50000', 'centerE6N4geo100000', 'centerE6N4geo150000',
    'uniformgeo',
    'centerX19Y4pca0.0015', 'centerX19Y4pca0.0025', 'centerX19Y4pca0.005',
    'uniformpca'
]

# Reorder the means and standard deviations based on custom order
means = means[custom_order]
std_devs = std_devs[custom_order]

# Create a plot with error bars
plt.figure(figsize=(8, 4))
plt.errorbar(means.index, means, yerr=std_devs, fmt='o', capsize=5, color='b', ecolor='black', elinewidth=2, markerfacecolor='black')
plt.xticks(rotation=90)
plt.title("Missense")

# Calculate and print the percentage change from the leftmost to rightmost value
percentage_change = ((means.iloc[-1] - means.iloc[0]) / means.iloc[0]) * 100
print(f"Percentage change from leftmost to rightmost value: {percentage_change:.4f}%")
print(means)
plt.ylim(0.175,0.176)
```

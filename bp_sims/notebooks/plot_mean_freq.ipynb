```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import glob
import re
import sys
sys.path.append("../source")
from plot_sim_sfs import *
```

## Uniform - first moment

```python
files_n1 = glob.glob("../results/uniform/*n1_*.sfs")
df = pd.read_csv(files_n1[0],header=None)
data = np.zeros((len(files_n1),5))
pattern = r"s([\d.]+)_n([\d.]+)_mu([\d.e-]+)_rho([\d.]+)_L"
for i,f in enumerate(files_n1):
    match = re.search(pattern, f)
    s = float(match.group(1))
    n = float(match.group(2))
    mu = float(match.group(3))
    rho = float(match.group(4))
    df = pd.read_csv(f,header=None)
    k1_n1 = df.loc[1,0]
    data[i,:] = [mu,s,n,rho,k1_n1]
data = pd.DataFrame(data,columns=["mu","s","n","rho","k1_n1"])
data['mu_over_s'] = data['mu']/data['s']
```

```python
fig,ax = plt.subplots(figsize=(8,8))

unique_rho_values = sorted(data['rho'].unique())
colors = ['red','blue','green']
# colors = plt.cm.Dark2(np.linspace(0, 1, len(unique_rho_values)))
color_dict = dict(zip(unique_rho_values, colors))

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylabel('simulation value: n=1, k=1')
ax.set_xlabel(r'theory value: $\mu$/$s$')
ax.loglog(np.logspace(-7,-1),np.logspace(-7,-1),color='black')

for rho_value, color in color_dict.items():
    df_rho = data[data['rho'] == rho_value]
    ax.scatter(df_rho['mu_over_s'], df_rho['k1_n1'], color=color, label=f'rho={rho_value}', marker='X', s=100, alpha=0.7)

ax.legend()

# ax.plot(data['mu_over_s'],data['k1_n1'],marker='X',color='red',markersize=8,linestyle='')
ax.set_title("First moment; Uniform sampling; L=50")
plt.show()
```

## Uniform: second moment

```python
files_n2 = glob.glob("../results/uniform/*n2_*.sfs")
data = np.zeros((len(files_n2),6))
pattern = r"s([\d.]+)_n([\d.]+)_mu([\d.e-]+)_rho([\d.]+)_L([\d.]+)"
for i,f in enumerate(files_n2):
    match = re.search(pattern, f)
    s = float(match.group(1))
    n = float(match.group(2))
    mu = float(match.group(3))
    rho = float(match.group(4))
    L = float(match.group(5))
    df = pd.read_csv(f,header=None)
    k2_n2 = df.loc[2,0]
    N = rho*L*L
    second_moment = mu*N*(s*N+n)/(n*(s*N)**2)+mu**2/s**2
    data[i,:] = [mu,s,n,rho,k2_n2,second_moment]
data = pd.DataFrame(data,columns=["mu","s","n","rho","k2_n2","second_moment"])
fig,ax = plt.subplots(figsize=(8,8))

unique_rho_values = sorted(data['rho'].unique())
colors = ['red','blue','green']
color_dict = dict(zip(unique_rho_values, colors))

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylabel('simulation value: n=2, k=2')
ax.set_xlabel(r'theory value')
ax.loglog(np.logspace(-9,-1),np.logspace(-9,-1),color='black')

for rho_value, color in color_dict.items():
    df_rho = data[data['rho'] == rho_value]
    ax.scatter(df_rho['second_moment'], df_rho['k2_n2'], color=color, label=f'rho={rho_value}', marker='X', s=100, alpha=0.7)

ax.legend()

ax.set_title("Second moment; Uniform sampling; L=50")
plt.show()
```

## Gaussian: first moment

```python
data = pd.read_csv("../../theory/old_files/results/spatial_integrals_dim2.csv")
data_pr = pd.read_csv("../../theory/old_files/results/cleaned_data_dim2_errorFalse.csv")
data_pr = data_pr.loc[data_pr['poly_type']=='1_1']
w_vals = data['w'].tolist()
res_vals = data_pr['residues']
pole_vals = data_pr['poles']
```

```python
files_n1 = glob.glob("../results/gaussian_w*/*n1_*L50_*iter100000000*.sfs")
data = np.zeros((len(files_n1),9))
pattern = r's([\d.]+)_n(\d+)_mu([\de.-]+)_rho(\d+)_L(\d+)_maxcount(\d+)_sigma([\d.]+)_seed(\d+)_iter(\d+)_r([\d.]+)_gaussian_w(\d+)\.sfs'
for i,f in enumerate(files_n1):
    match = re.search(pattern, f)
    s = float(match.group(1))
    n = float(match.group(2))
    mu = float(match.group(3))
    rho = float(match.group(4))
    sigma = float(match.group(7))
    w = float(match.group(11))
    df = pd.read_csv(f,header=None)
    k1_n1 = df.loc[1,0]
    theta_e = get_thetae(w,s,w_vals,res_vals,mu,rho,sigma,d=2)
    gamma_e = get_gammae(w,s,w_vals,pole_vals,rho,sigma,d=2)
    data[i,:] = [theta_e,gamma_e,n,rho,k1_n1,w,s,mu,sigma]

df = pd.DataFrame(data,columns=["theta_e","gamma_e","n","rho","k1_n1","w","s","mu","sigma"])
df['theta_over_gamma'] = df['theta_e']/df['gamma_e']
```

```python
unique_rho_values = sorted(df['rho'].unique())
colors = ['red','blue','green']
color_dict = dict(zip(unique_rho_values, colors))

fig,ax = plt.subplots(figsize=(8,8))
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylabel('simulation value: n=1, k=1')
ax.set_xlabel(r'theory value: $\theta_E$/$\gamma_E$')
ax.loglog(np.logspace(-8,-1),np.logspace(-8,-1),color='black')

marker_shapes = ["^",'+','x']

for i, rho_value in enumerate(sorted(df['rho'].unique())):
    df_rho = df[df['rho'] == rho_value]
    for j, w_value in enumerate(sorted(df_rho['w'].unique())):
        df_w = df_rho[df_rho['w'] == w_value]
        ax.scatter(df_w['theta_over_gamma'], df_w['k1_n1'], color=color_dict[rho_value], label=f'rho={rho_value}, w={w_value}', marker=marker_shapes[j], s=100, alpha=0.7)

ax.legend()
ax.set_title("First moment; Gaussian sampling; L=50")
plt.show()
```

## Gaussian: second moment

```python
files_n2 = glob.glob("../results/gaussian_w*/*n2_*L50_*iter100000000*.sfs")
data = np.zeros((len(files_n2),7))
pattern = r's([\d.]+)_n(\d+)_mu([\de.-]+)_rho(\d+)_L(\d+)_maxcount(\d+)_sigma([\d.]+)_seed(\d+)_iter(\d+)_r([\d.]+)_gaussian_w(\d+)\.sfs'
for i,f in enumerate(files_n2):
    match = re.search(pattern, f)
    s = float(match.group(1))
    n = float(match.group(2))
    mu = float(match.group(3))
    rho = float(match.group(4))
    sigma = float(match.group(7))
    w = float(match.group(11))
    df = pd.read_csv(f,header=None)
    k2_n2 = df.loc[2,0]
    theta_e = get_thetae(w,s,w_vals,res_vals,mu,rho,sigma,d=2)
    gamma_e = get_gammae(w,s,w_vals,pole_vals,rho,sigma,d=2)
    second_moment = (theta_e*(gamma_e+n))/(n*gamma_e**2)+(theta_e**2/gamma_e**2)
    data[i,:] = [theta_e,gamma_e,n,rho,k2_n2,w,second_moment]

df = pd.DataFrame(data,columns=["theta_e","gamma_e","n","rho","k2_n2","w","second_moment"])
```

```python
fig,ax = plt.subplots(figsize=(8,8))
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylabel('simulation value: n=2, k=2')
ax.set_xlabel(r'theory value: $\mathbb{E}[p^2]$')
ax.loglog(np.logspace(-8,-1),np.logspace(-8,-1),color='black')
marker_shapes = ["^",'+','x']

for i, rho_value in enumerate(sorted(df['rho'].unique())):
    df_rho = df[df['rho'] == rho_value]
    for j, w_value in enumerate(sorted(df_rho['w'].unique())):
        df_w = df_rho[df_rho['w'] == w_value]
        ax.scatter(df_w['second_moment'], df_w['k2_n2'], color=color_dict[rho_value], label=r'$\rho$='+str(rho_value)+r', $w$='+str(w_value), marker=marker_shapes[j], s=100, alpha=0.7)

ax.legend()
ax.set_title("Second moment; Gaussian sampling; L=50")
plt.show()
```

## Proportion of segregating sites

```python
files_n100 = glob.glob("../results/gaussian_w*/*n100_*L50_*iter100000000*.sfs")
data = np.zeros((len(files_n100),7))
# pattern = r's(\d+)_n(\d+)_mu([\d.e-]+)_rho([\d.e-]+)_L(\d+)_maxcount(\d+)_sigma([\d.e-]+)_seed(\d+)_iter(\d+)_r([\d.e-]+)_gaussian_w(\d+)\.sfs'
pattern = r's([\d.]+)_n(\d+)_mu([\de.-]+)_rho(\d+)_L(\d+)_maxcount(\d+)_sigma([\d.]+)_seed(\d+)_iter(\d+)_r([\d.]+)_gaussian_w(\d+)\.sfs'
for i,f in enumerate(files_n100):
#     print(f)
    match = re.search(pattern, f)
    s = float(match.group(1))
    n = float(match.group(2))
    mu = float(match.group(3))
    rho = float(match.group(4))
    sigma = float(match.group(7))
    w = float(match.group(11))
    df = pd.read_csv(f,header=None)
    prop_seg_sim = 1-df.loc[0,0]
    theta_e = get_thetae(w,s,w_vals,res_vals,mu,rho,sigma,d=2)
    gamma_e = get_gammae(w,s,w_vals,pole_vals,rho,sigma,d=2)
    prop_seg_theory = 1-(gamma_e/(gamma_e+n))**theta_e
    data[i,:] = [theta_e,gamma_e,n,rho,w,prop_seg_sim,prop_seg_theory]

df = pd.DataFrame(data,columns=["theta_e","gamma_e","n","rho","w","prop_seg_sim","prop_seg_theory"])
```

```python
fig,ax = plt.subplots(figsize=(8,8))
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylabel('simulation value')
ax.set_xlabel(r'theory value')
ax.loglog(np.logspace(-7,0),np.logspace(-7,0),color='black')

marker_shapes = ["^",'+','x']

for i, rho_value in enumerate(sorted(df['rho'].unique())):
    df_rho = df[df['rho'] == rho_value]
    for j, w_value in enumerate(sorted(df_rho['w'].unique())):
        df_w = df_rho[df_rho['w'] == w_value]
        ax.scatter(df_w['prop_seg_theory'], df_w['prop_seg_sim'], color=color_dict[rho_value], label=r'$\rho$='+str(rho_value)+r', $w$='+str(w_value), marker=marker_shapes[j], s=100, alpha=0.7)

ax.legend()
ax.set_title("Prop. segregating sites; Gaussian sampling; L=50; n=100")
plt.show()
```
